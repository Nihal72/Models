```python
columns = ['item_id', 'movie title', 'release data', 'video release date', 'IMDb URL', 'unknown', 'Acrion', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentry' , 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci - Fi', 'Thriller', 'War', 'Western']
```

```python
movies = pd.read_csv('ml-100k/u.item',error_bad_lines= False, encoding = 'latin-1', sep = '|', names = columns)
movies
```

```python
movie_names = movies[['item_id', 'movie title']]
movie_names.head()
```

```python
frame.head()
```

```python
combined_movies_data = pd.merge(frame, movie_names , on = 'item_id')
combined_movies_data.head()
```

```python
combined_movies_data.groupby('item_id')['rating'].count().sort_values(ascending = False).head()
```

```python
filter = combined_movies_data['item_id'] == 50
combined_movies_data[filter]['movie title'].unique()
```

```python
rating_crosstab = combined_movies_data.pivot_table(values = 'rating', index = 'user_id', columns = 'movie title', fill_value= 0)
```

```python
rating_crosstab.head()
```

```python
rating_crosstab.describe()
```

```python
rating_crosstab.shape
```

```python
# Transpose(We need movies names as a recomendation)
X = rating_crosstab.T
X.shape
```

```python
#Decomposing the Matrix
from sklearn.decomposition import TruncatedSVD
SVD = TruncatedSVD(n_components = 12, random_state = 17)
resultant_matrix = SVD.fit_transform(X)
resultant_matrix.shape
```

```python
resultant_matrix
```

```python
#Generating a correaltion Matrix
corr_mat = np.corrcoef(resultant_matrix)
corr_mat
```

```python
corr_mat.shape
```

```python
# Isolating star wars from the correlation matrix

movies_names = rating_crosstab.columns
movies_names
```

```python
corr_mat.shape
```

```python
movies_list = list(movies_names)
movies_list
```

```python
star_wars = movies_list.index('Star Wars (1977)')
star_wars
```
